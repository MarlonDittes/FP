```python
import openpyxl

def read_data_from_file(file_path):
    with open(file_path, 'r') as file:
        data = [line.strip().split() for line in file]

    return data

# Example file paths
file_paths = ['performance_data/1', 'performance_data/2', 'performance_data/3', 'performance_data/4']  # Replace with your actual file paths

# Create a new Excel workbook and select the active sheet
workbook = openpyxl.Workbook()
sheet = workbook.active

# Add headers to the sheet
headers = ["Graph"]
for file_index, file_path in enumerate(file_paths, start=1):
    if file_index == 1:
        name = "Brute Force"
    elif file_index == 2:
        name = "Greedy"
    elif file_index == 3:
        name = "Baricenter"
    elif file_index == 4:
        name = "Baricenter + Greedy"

    headers.extend([f"{name} Duration [ns]", f"{name} Crossings"])

sheet.append(headers)

for file_index, file_path in enumerate(file_paths, start=1):
    # Read data from the file
    data = read_data_from_file(file_path)

    # Extract test name, duration, and crossings from the data
    for row_data in data:
        test_name, duration, crossings = row_data
        header = test_name

        # Find the row for the current test name
        test_name_row = None
        for row in sheet.iter_rows(min_row=2, max_row=sheet.max_row, min_col=1, max_col=1):
            if row[0].value == header:
                test_name_row = row[0].row
                break

        # If the test name is not found, add a new row
        if test_name_row is None:
            test_name_row = sheet.max_row + 1
            sheet.cell(row=test_name_row, column=1, value=header)

        # Populate the corresponding columns with data
        sheet.cell(row=test_name_row, column=file_index * 2, value=duration)
        sheet.cell(row=test_name_row, column=file_index * 2 + 1, value=crossings)

# Save the workbook to a file
workbook.save("combined_results.xlsx")
```
